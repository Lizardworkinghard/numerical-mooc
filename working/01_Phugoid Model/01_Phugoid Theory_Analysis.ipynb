## Understanding the Equation

Having derived the equations, there are still some detailes unclear regarding the constant $C$ in the Phugoid equation. It is clear that it cannot be larger than $2/3$, however, the fact that it can be negative is somehow confusing.
This condition is reached when the relationship between the angle $\theta$ and the ratio $\phi$ is such that $\theta \leq \arccos(1/3\phi)$. The equation for C was obtained to be:

$$\begin{equation}
C = \dfrac{\cos(\theta)-\frac{1}{3\phi}}{\sqrt{\phi}}
\end{equation}$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
"Discretisation for theta. We consider theta from 0 to 2pi, with a step of pi/4, i.e., there will be 9 values for theta."
delta_theta = np.pi/4
theta = np.linspace(0, 2*np.pi, (2*np.pi/delta_theta)+1)
print(theta)
```

```python
"Definition of the range for phi. We take the starting poitn at 0.01 in order to avoid dividing by zero"
phi = np.linspace(0.01,20,100)
```

```python
"Empty array and definition of its shape"
c = np.empty([9,100])
```

```python
"Each one fo the 9 rows in C correspond to C in function of phi for a given value of theta"
for i in range(0,9):
    c[i] = (np.cos(theta[i])-1/(3*phi))/phi**0.5
```

```python
for i in range(0,5):
    plt.plot(phi,c[i])
    plt.axis([-0.5,15,-4,1])
```

It can be seen how $C$ can be any negative number. The graph only shows the curves for $\theta = 0, \pi/4, \pi/2, 3\pi/4, \pi$, which corespond to half the period of the cosine function.
It is showed how for $\pi/2 \leq \theta \leq 3\pi/2$ $C$ is always negative, while for the rest values of $\theta$ the sign of $C$ depends on the value of $\phi$ as well.

Now we can confirm that the frequency of the oscillation increases when the heights increases. We can now confirm that the amplitudes are different as well. The path starts at $z$ at the angle $\theta$. The equilibrium level is set at $z_t$, so the phugoid oscillates around it.

After analysing the graphs in the course material, it is clear that the specified $\theta$ and $z$ are the initial conditions of the path. We already said that $z_t$ is the reference or equilibrium.
The increasing "perceived" frequency in the plotted paths is explained by considering the phugoid equation in terms of $R$, $z$ and $z_t$:

$$\begin{eqnarray}
\dfrac{z_t}{R} = \dfrac{1}{3} - C\sqrt{\phi^3} \\
R = \dfrac{z_t}{\frac{1}{3}-C\sqrt{\phi^3}}
\end{eqnarray}$$

And since $C$ and $\phi$ remained constant, the change in $R$ was due to $\delta = z_t - z$. The larger $\delta$, the larger $R$ so more space (x axis) is required for the phugoid to complete an oscillation.
